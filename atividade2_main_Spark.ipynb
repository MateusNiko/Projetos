```
# Cmd 16 - Transformação dos dados removendo as vírgulas e convertendo para float e timestamp
```

```
df_without_coma = spark.sql("select cast(replace(cotacaoCompra, ',', '.') as float) as cotacaoCompra, cast(replace(cotacaoVenda, ',', '.') as float) as cotacaoVenda, to_timestamp(dataHoraCotacao) as dataHoraCotacao from Tabela").collect()
```

```
# Cmd 20 - Criação de um DataFrame com os dados da lista df_without_coma
```

```
final_data = spark.createDataFrame(df_without_coma)
```

```
# Cmd 22 - Visualização dos dados do novo DataFrame
```

```
final_data.show()
```

```
# Cmd 24 - Escrever o DataFrame no formato CSV no Blob Storage
```

```
final_data.coalesce(1).write.mode("overwrite").option("header", "true").format("com.databricks.spark.csv").save(f"wasbs://mateus-nikolofski@stgestudos.blob.core.windows.net/Atividade2_Spark")
```
