```
# Cmd 9 - Abrir o arquivo copiado e passar os dados no formato de lista.
```

```
with open("/tmp/copy_atividade2_mateus_nikolofski") as file:
    data = file.readlines()
```

```
# Cmd 11 - Declarar listas com nomes das colunas do arquivo CSV para manipular os dados e organizá-los em uma String final.
```

```
cotacaoCompra = []
cotacaoVenda = []
dataHoraCotacao = []
```

```
# Cmd 13 - Dividir a lista para somente ter os dados sem os headers.
```

```
content = data[2:]
```

```
# Cmd 15 - Remover as vírgulas e as aspas. Após isso, reagrupar os dados e colocá-los em listas diferentes.
# first_column representa os dados da cotacaoCompra
# middle_column representa os dados da cotacaoVenda
# last_column representa os dados da dataHoraCotacao
```

```
first_column = []
middle_column = []
last_column = []
for n in range(len(content)):
    cleaner = content[n].replace('"', "").split(',')

    first_column.append(f"{cleaner[0]}.{cleaner[1]}")
    middle_column.append(f"{cleaner[2]}.{cleaner[3]}")
    last_column.append(f"{cleaner[4].strip()}")




for data in first_column:
    cotacaoCompra.append(data)

for data in middle_column:
    cotacaoVenda.append(data)

for data in last_column:
    dataHoraCotacao.append(data)
```

```
# Cmd 23 - csv_list é uma lista que agrupa as outras listas juntas, já preparando os dados para serem formatados em CSV.
```

```
csv_list = ["cotacaoCompra", "cotacaoVenda", "dataHoraCotacao"]         
```

```
# Cmd 25 - Iterador for para agrupar os itens de outras listas em uma única.
```

```
for i in range(len(cotacaoCompra)):
    csv_list.append(cotacaoCompra[i])
    csv_list.append(cotacaoVenda[i])
    csv_list.append(dataHoraCotacao[i])
```

```
# Cmd 27 - Formação da String para ser utilizada como parâmetro na função que reescreve o arquivo CSV do Blob Storage.
```

```
string_a_csv = ""
counter = 0
for i in range(len(csv_list)):
    if counter <= 1:
        string_a_csv += csv_list[i] + ","
        counter += 1
    else:
        string_a_csv += csv_list[i]
        counter += 1
    if counter == 3:
        string_a_csv +="\n"
        counter = 0
```

```
# Cmd 29 - Função do Databricks utilities para reescrever o arquivo no Blob Storage. Nela são solicitados o local do arquivo no Databricks e uma String.
```

```
dbutils.fs.put("/mnt/atividade2_mateus_nikolofski", string_a_csv, True)
```
